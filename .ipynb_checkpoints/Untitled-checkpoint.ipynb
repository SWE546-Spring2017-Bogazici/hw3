```python
from sklearn.datasets import load_iris
from sklearn import tree
from IPython.display import Image  
import pydotplus as py

def DecisionTree(data, target, criterion, depth):
    clf = tree.DecisionTreeClassifier(criterion = criterion, max_depth = depth)
    clf = clf.fit(data, target)
    return clf
```

```python
# Decision tree with gini index

irisdata = load_iris()

clf_gini = DecisionTree(irisdata.data, irisdata.target, "gini", 18)
dot_data = tree.export_graphviz(clf_gini, out_file=None, feature_names = irisdata.feature_names
                                , class_names = irisdata.target_names, filled=True, rounded=True
                                , special_characters=True) 
graph = py.graph_from_dot_data(dot_data) 
Image(graph.create_png())
```

```python
# Decision tree with entropy index

clf_entr = DecisionTree(irisdata.data, irisdata.target, "entropy", 10)
dot_data = tree.export_graphviz(clf_entr, out_file=None, feature_names = irisdata.feature_names
                                , class_names = irisdata.target_names, filled=True, rounded=True
                                , special_characters=True) 
graph = py.graph_from_dot_data(dot_data) 
Image(graph.create_png())
```

```python
# Finding classification accuracy for both gini and entropy indexes

from sklearn.model_selection import LeaveOneOut

def DecisionTreeAccuracy (data, target, index, depth):
    success = 0
    totalnum = 0
    for train, test in LeaveOneOut().split(data):        
        loo_data = list()
        loo_target = list()
        for i in train:
            loo_data.append(data[i])
            loo_target.append(target[i])

        clf = DecisionTree(loo_data, loo_target, index, depth)
        totalnum = totalnum + 1
        if (target[test]==clf.predict(data[test] )):
            success = success + 1
    return (success / totalnum)  


```

```python
for i in range (1,10):
    print("For depth " + str(i) + " and gini index success rate is % " 
           + str(DecisionTreeAccuracy(irisdata.data, irisdata.target, "gini", i)*100))
    print("For depth " + str(i) + " and entropy index success rate is % " 
           + str(DecisionTreeAccuracy(irisdata.data, irisdata.target, "entropy", i)*100))
```
